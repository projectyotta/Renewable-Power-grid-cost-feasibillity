```python
# now , let us put all of this into a structured format 
# create a list for each item that we want 
list_date=[] 
list_open=[]
list_high=[] 
list_low =[]
list_close=[] 
list_volume=[] 
list_marketcap=[]

for x in range(0,len(divclass2)):
    divclass4=divclass2[x].find_all("td")
    s1=str(divclass4[0])
    s1=str(divclass4[0])
    s2=str(divclass4[1])
    s3=str(divclass4[2])
    s4=str(divclass4[3])
    s5=str(divclass4[4])
    s6=str(divclass4[5])
    s7=str(divclass4[6])
    list_date.append(s1[s1.find(start_fordate)+len(start_fordate):s1.rfind(end)])
    list_open.append(s2[s2.find(start_normal)+len(start_normal):s2.rfind(end)])
    list_high.append(s3[s3.find(start_normal)+len(start_normal):s3.rfind(end)])
    list_low.append(s4[s4.find(start_normal)+len(start_normal):s4.rfind(end)])
    list_close.append(s5[s5.find(start_normal)+len(start_normal):s5.rfind(end)])
    list_volume.append(s6[s6.find(start_normal)+len(start_normal):s6.rfind(end)])
    list_marketcap.append(s7[s7.find(start_normal)+len(start_normal):s7.rfind(end)])
    
    
print("done")
```

```python
# put all of these lists in a single dataframe
finaldf = pd.DataFrame(
    {'Date': list_date,
     'Open': list_open,
     'High': list_high, 
     'Low' : list_low, 
     'Close' : list_close, 
     'Volume' : list_volume, 
     'MarketCap' : list_marketcap
    })
print("done")
```
